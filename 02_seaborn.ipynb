## ■(1) 準備

```python
# データ操作とグラフ描画のライブラリ
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## ■(2) データの確認

```python
# タイタニックのデータ読み込み
df = sns.load_dataset('titanic')
df

# survived：生存フラグ（0=死亡、1=生存）
# pclass：チケットクラス（1=上層、2=中層、3=下層）
# sex：性別（male=男性、female＝女性）
# age：年齢
# sibsp：タイタニックに同乗している兄弟/配偶者の数
# parch：タイタニックに同乗している親/子供の数
# fare：料金
# embarked：出港地（タイタニックへ乗った港 C=Cherbourg、Q=Queenstown、S=Southampton）
# class：クラス（1=First、2=Second、3=Third）
# who：男性(man) or 女性(woman)
# adult_male：成人男性であるかどうか
# deck：乗船していたデッキ
# embark_town：出港地
# alive：生存したかどうか
# alone：一人であったかどうか
```

## ■(3) 棒グラフ

```python
# 棒グラフの描画(生存フラグ)
sns.catplot(x="survived", data=df, kind="count")
plt.show()
```

## ■(4) ヒストグラム

```python
# ヒストグラムの描画(年齢)
sns.displot(df['age'])
plt.show()
```

## ■(5) 箱ひげ図

```python
# 箱ひげ図の描画(年齢と性別による生存数の対比)
sns.catplot(x='sex', y='age', data=df, kind='box', hue='survived')
plt.show()
```

## ■(6) ストリッププロットの描画

```python
# ストリッププロットの描画(年齢と性別による生存数の対比)
sns.catplot(x='sex', y='age', data=df, hue='survived')
plt.show()
```

## ■(7) ヴァイオリンプロット

```python
# ヴァイオリンプロットの描画(年齢と性別による生存数の対比)
sns.catplot(x='sex', y='age', data=df, kind='violin', hue='survived')
plt.show()
```

## ■(8) 散布図行列

```python
# アヤメのデータの読み込み
iris = sns.load_dataset('iris')
iris

# sepal_length：がくの長さ
# sepal_width ：がくの幅
# petal_length：花弁の長さ
# petal_width ：花弁の幅
# species     ：アヤメの種類
```

```python
# 散布図行列の描画(がくの長さと幅、花弁の長さと幅の組み合わせ)
sns.pairplot(iris)
plt.show()
```

```python
# 種類(species)別の色分け
sns.pairplot(data=iris, hue='species')
plt.show()
```

## ■(9) ヒートマップ

```python
# ヒートマップの描画(昇順)
list = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
sns.heatmap(list)
plt.show()
```

```python
# ヒートマップの描画(降順)
list = [[8, 7, 6],[5, 4, 3],[2, 1, 0]]
sns.heatmap(list)
plt.show()
```
